```python
import cv2
import numpy as np

# 读取灰度图像
image_path = 'newleaflet.jpg'
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# 应用中值滤波来减少噪点
denoised_image = cv2.medianBlur(image, 3)

# 使用自适应阈值进行二值化
binary_image = cv2.adaptiveThreshold(denoised_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                     cv2.THRESH_BINARY, 11, 2)

# 形态学操作：腐蚀和膨胀，减少噪音
kernel = np.ones((2, 2), np.uint8)
morph_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel)

# 保存处理后的图像
cv2.imwrite('newleaflet_pre.jpg', morph_image)

# 显示处理后的图像
cv2.imshow('newleaflet_pre', morph_image)
cv2.waitKey(0)
cv2.destroyAllWindows()

```

```python
import pytesseract
import cv2
import os

pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

image_path = 'newleaflet.jpg' 
image = cv2.imread(image_path)

image_name = os.path.splitext(os.path.basename(image_path))[0]

boxes = pytesseract.image_to_boxes(image, lang='jpn+chi_tra', config='--psm 6')

height, width, _ = image.shape

output_dir = f'{image_name}_seg'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

page_number = 1  
index = 0  

padding = 10  

for b in boxes.splitlines():
    b = b.split(' ')
    char, x, y, w, h = b[0], int(b[1]), int(b[2]), int(b[3]), int(b[4])

    y = height - y
    h = height - h

    x = max(0, x - padding)
    w = min(width, w + padding)
    h = max(0, h - padding)
    y = min(height, y + padding)

    char_image = image[h:y, x:w]

    char_image_name = f"{image_name}_{page_number}_{index}.jpg"

    cv2.imwrite(os.path.join(output_dir, char_image_name), char_image)

    index += 1
```
